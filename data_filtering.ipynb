```python
import pandas as pd
```

```python
df = pd.read_csv('database/base_experimental.csv')
df.head()
```

## Data processing
### Convert values to propper python types

```python
from ast import literal_eval
df['keywords'] = df['keywords'].fillna('[]').transform(literal_eval)
df['authors'] = df['authors'].fillna('[]').transform(literal_eval)
df['references'] = df['references'].fillna('[]').transform(literal_eval)
df['abstract'] = df['abstract'].fillna('')
df.head()
```

### Convert all words in *keywords*, *title* and *abstract* to lowercase

```python
df['keywords'] = df['keywords'].apply(lambda x: [i.lower() for i in x])
df['abstract'] = df['abstract'].str.lower()
df['title'] = df['title'].str.lower()
df.head()
```

### Leave only journal articles

```python
df_filtered = df[df['type'] == 'Journal Article']
```
